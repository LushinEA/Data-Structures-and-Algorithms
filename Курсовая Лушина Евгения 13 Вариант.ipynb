```python
class Deque:
    def __init__(self):
        self.items = []
    
    def push_right(self, item): # –¥–æ–±–∞–≤–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç —Å–ø—Ä–∞–≤–∞
        self.items.append(item)
        
    def push_left(self, item): # –¥–æ–±–∞–≤–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç —Å–ª–µ–≤–∞
        self.items.insert(0, item)
    
    def pop_right(self): # —É–¥–∞–ª–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç —Å–ø—Ä–∞–≤–∞
        return self.items.pop()
    
    def pop_left(self): # —É–¥–∞–ª–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç —Å–ª–µ–≤–∞
        return self.items.pop(0)
    
    def top_right(self): # –ø—Ä–æ—Å–º–æ—Ç—Ä –ø—Ä–∞–≤–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
        return self.items[len(self.items) - 1]
    
    def top_left(self): # –ø—Ä–æ—Å–º–æ—Ç—Ä –ª–µ–≤–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
        return self.items[0]
    
    def empty(self): # –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –ø—É—Å—Ç–æ—Ç—É –¥–µ–∫–∞
        return self.items == []
    
    def size(self): # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –¥–µ–∫–µ
        return len(self.items)

print("-----------------------------–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ-----------------------------")
n = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω: ")) # –í–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω
list_data = [] # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –¥–ª—è –∑–∞–ø–∏—Å–∏ –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π degree –∏ s
single_vertex = Deque() # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º Deque
for i in range(n): # –¶–∏–∫–ª –æ—Ç 0 –¥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–µ—Ä—à–∏–Ω
    degreev, sv = map(int, input("–í–≤–µ–¥–∏—Ç–µ degreev –∏ sv: ").split()) # –í–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–º–µ–∂–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω –∏ XOR —Ñ—É–Ω–∫—Ü–∏–π
    if (degreev == 1): # –ï—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–º–µ–∂–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω = 1, —Ç–æ –º–æ–∂–µ–º —Å—Ä–∞–∑—É –∑–∞–Ω–µ—Å—Ç–∏ –≤ –æ—á–µ—Ä–µ–¥—å
        single_vertex.push_right(i) # –ó–∞–Ω–æ—Å–∏–º –≤ –æ—á–µ—Ä–µ–¥—å
    list_data.append([degreev, sv]) # –ó–∞–ø–∏—Å—ã–≤–∞–µ–º –≤–≤–µ–¥—ë–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å–ø–∏—Å–æ–∫

count = 0 # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á—ë—Ç—á–∏–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Ä—ë–±–µ—Ä
edges = '' # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –≤—ã–≤–æ–¥–∞ –Ω–∞–∑–≤–∞–Ω–∏–π —Ä—ë–±–µ—Ä
while not single_vertex.empty(): # –ü–æ–∫–∞ –æ—á–µ—Ä–µ–¥—å –Ω–µ –ø—É—Å—Ç–∞
    vertex = single_vertex.pop_left() # –†–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º –≤–µ—Ä—à–∏–Ω—É –∏–∑ –æ—á–µ—Ä–µ–¥–∏
    degreev, sv = list_data[vertex] # –ë–µ—Ä—ë–º –∑–∞–ø–∏—Å–∞–Ω–Ω—É—é –ø–∞—Ä—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–º–µ–∂–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω –∏ XOR —Å—É–º–º—ã –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤–µ—Ä—à–∏–Ω—ã
    if degreev == 0: # –ï—Å–ª–∏ —Å–º–µ–∂–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω –Ω–µ—Ç, –ø—Ä–æ–ø—É—Å–∫–∞–µ–º –≤–µ—Ä—à–∏–Ω—É
        continue 
    second_degreev, second_sv = list_data[sv] # –ë–µ—Ä—ë–º –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ—Ä—à–∏–Ω—ã —Ä–∞–≤–Ω—è—é—â–µ–π—Å—è XOR —Å—É–º–º–µ
    if second_degreev == 2: # –ï—Å–ª–∏ —É –Ω–µ—ë —Å–º–µ–∂–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω 2
        single_vertex.push_right(sv) # –î–æ–±–∞–≤–ª—è–µ–º –≤ –æ—á–µ—Ä–µ–¥—å
    list_data[sv] = [second_degreev-1, second_sv^vertex] # –£–¥–∞–ª—è–µ–º —Ä–µ–±—Ä–æ
    edges += str(vertex)+' '+str(sv)+'\n' # –ó–∞–ø–∏—Å—ã–≤–∞–µ–º –Ω–∞–π–¥–µ–Ω–Ω–æ–µ —Ä–µ–±—Ä–æ
    count += 1 # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á—ë—Ç—á–∏–∫ —Ä—ë–±–µ—Ä
print("-----------------------------–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ----------------------------")
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—ë–±–µ—Ä –≥—Ä–∞—Ñ–∞ = ", count) # –í—ã–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—ë–±–µ—Ä
print("–†—ë–±—Ä–∞")
print(edges) # –í—ã–≤–æ–¥–∏–º —Å–∞–º–∏ —Ä—ë–±—Ä–∞
```

```python
print("-----------------------------–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ-----------------------------")
n, m, k = map(int, input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ö–æ–±–±–∏—Ç–æ–≤, –æ–±—â–µ–µ —á–∏—Å–ª–æ –ø–æ–¥—É—à–µ–∫, ‚Ññ –∫—Ä–æ–≤–∞—Ç–∏ –§—Ä–æ–¥–æ, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ: ").split()) # –í–≤–æ–¥ -> 
     #-> –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Ö–æ–±–±–∏—Ç–æ–≤, –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–¥—É—à–µ–∫, –Ω–æ–º–µ—Ä –∫—Ä–æ–≤–∞—Ç–∏ –§—Ä–æ–¥–æ

def one_sided(slots, x): # –§—É–Ω–∫—Ü–∏—è –ø–æ–¥—Å—á—ë—Ç–∞ –ø–æ–¥—É—à–µ–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫—Ä–∞—è –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞
    if slots >= x-1: # –ï—Å–ª–∏ –∫—Ä–æ–≤–∞—Ç–µ–π –±–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ –æ—Å—Ç–∞–≤—à–∏–º—Å—è –ø–æ–¥—É—à–∫–∞–º
        result = (slots - x + 1) + (x-1)* x // 2
        return result
    else:
        result = slots * ( 2* x - slots -1 ) // 2
        return result

def pillows_used(x): # –§—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –ø–æ–¥—É—à–µ–∫
    left_slot = k -1 # –ö—Ä–æ–≤–∞—Ç–µ–π —Å–ª–µ–≤–∞ –æ—Ç –§—Ä–æ–¥–æ
    right_slot = n-k # –ö—Ä–æ–≤–∞—Ç–µ–π —Å–ø—Ä–∞–≤–∞ –æ—Ç –§—Ä–æ–¥–æ
    res = x + one_sided(left_slot, x) + one_sided(right_slot, x) <= m # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –µ—â—ë —Å–≤–æ–±–æ–¥–Ω—ã–µ –ø–æ–¥—É—à–∫–∏ –∏–ª–∏ ->
                                                                      #-> –ø–æ–¥—É—à–µ–∫ –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å –±–æ–ª—å—à–µ —á–µ–º –µ—Å—Ç—å
    return res

left = 1 # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –¥–ª—è –ª–µ–≤–æ–π –≥—Ä–∞–Ω—Ü–∏—ã (–∑–∞ –≥—Ä–∞–Ω–∏—Ü—É –±–µ—Ä—ë–º –ø–æ–¥—É—à–∫–∏)
right = m # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –¥–ª—è –ø—Ä–∞–≤–æ–π –≥—Ä–∞–Ω–∏—Ü—ã
mid = (left + right) // 2 # –ù–∞—Ö–æ–¥–∏–º —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ–¥—É—à–µ–∫ (–ø–æ —Å—É—Ç–∏ –∫–æ–ª-–≤–æ –ø–æ–¥—É—à–µ–∫ –§—Ä–æ–¥–æ +1)
while left != right: # –ü–æ–∫–∞ –≥—Ä–∞–Ω–∏—Ü—ã –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–ª–∏ –¥—Ä—É–≥ –¥—Ä—É–≥–∞
    if pillows_used(mid): # –ï—Å–ª–∏ –æ—Å—Ç–∞—é—Ç—Å—è –Ω–µ–≤–æ—Å—Ç—Ä–µ–±–æ–≤–∞–Ω–Ω—ã–µ –ø–æ–¥—É—à–∫–∏
        left = mid + 1 # –¢–æ –Ω–∏–∂–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É –Ω—É–∂–Ω–æ —É–≤–µ–ª–∏—á–∏—Ç—å (—É–≤–µ–ª–∏—á–∏–≤–∞–µ–º –ø–æ —Ñ–∞–∫—Ç—É –∫–æ–ª-–≤–æ –ø–æ–¥—É—à–µ–∫ —É –§—Ä–æ–¥–æ –∏ –ø–æ –æ—Å—Ç–∞—Ç–æ—á–Ω–æ–º—É –ø—Ä–∏–Ω—Ü–∏–ø—É —É —Ö–æ–±–±–∏—Ç–æ–≤)
    else: # –ò–Ω–∞—á–µ –µ—Å–ª–∏ –ø–æ–¥—É—à–µ–∫ –Ω–µ —Ö–≤–∞—Ç–∏–ª–æ
        right = mid # –£–º–µ–Ω—å—à–∞–µ–º –≤–µ—Ä—Ö–Ω—é—é –≤–µ—Ä—Ö–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É (—É–º–µ–Ω—å—à–∞–µ–º –ø–æ —Ñ–∞–∫—Ç—É –∫–æ–ª-–≤–æ –ø–æ–¥—É—à–µ–∫ —É –§—Ä–æ–¥–æ –∏ –ø–æ –æ—Å—Ç–∞—Ç–æ—á–Ω–æ–º—É –ø—Ä–∏–Ω—Ü–∏–ø—É —É —Ö–æ–±–±–∏—Ç–æ–≤)
    mid = (left + right) // 2 # –û–±–Ω–æ–≤–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ–¥—É—à–µ–∫
print("-----------------------------–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ----------------------------")
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–¥—É—à–µ–∫ —É –§—Ä–æ–¥–æ = ", mid - 1) # –í—ã–≤–æ–¥–∏–º –∫–æ–ª-–≤–æ –ø–æ–¥—É—à–µ–∫ —É –§—Ä–æ–¥–æ
```

```python
print("-----------------------------–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ-----------------------------")
n, m = map(int, input("–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—É –ø—Ä–æ–≥—Ä–∞–º–º—ã –∏ –º–æ–º–µ–Ω—Ç –æ—Ç–∫–ª—é—á–µ–Ω–∏—è —Å–≤–µ—Ç–∞, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ: ").split()) # –í–≤–æ–¥ –¥–ª–∏–Ω—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã ->
                                                                                                #-> –∏ –∫–æ–º–∞–Ω–¥—ã –Ω–∞ –≤—ã–∫–ª—é—á–µ–Ω–∏–µ

program = [0] + [int(x) for x in input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–∞ –ø—Ä–æ–≥—Ä–∞–º–º—ã a: ").split()] + [m] # –î–æ–±–∞–≤–∏–º –∫ –º–∞—Å—Å–∏–≤—É –ø—Ä–æ–≥—Ä–∞–º–º—ã –≤ –Ω–∞—á–∞–ª–æ 0 ->
         #-> (–≤–∫–ª—é—á–µ–Ω–∏–µ –ø–∏—Ç–∞–Ω–∏—è –∏ —Å–≤–µ—Ç–∞) –∏ –∫–æ–º–∞–Ω–¥—ã –ú –≤ –∫–æ–Ω–µ—Ü –º–∞—Å—Å–∏–≤–∞ (–≤—ã–∫–ª—é—á–µ–Ω–∏–µ –ø–∏—Ç–∞–Ω–∏—è)
    
lamp_off = 0 # –°—É–º–º–∞—Ä–Ω–æ–µ –≤—Ä–µ–º—è –Ω–µ —Ä–∞–±–æ—Ç—ã –ª–∞–º–ø–æ—á–∫–∏
lamp_on = 0 # –°—É–º–º–∞—Ä–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –ª–∞–º–ø–æ—á–∫–∏
final = 0 # –í—ã–∏–≥—Ä–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è

for i in range(1, n+2): # –û—Ç 1 –Ω–æ–º–µ—Ä–∞ –º–∞—Å—Å–∏–≤–∞ (–æ—Ç 0 –µ—Å–ª–∏ –Ω–µ —Å—á–∏—Ç–∞—Ç—å –¥–æ–±–∞–≤–ª–µ–Ω–Ω—ã–π –Ω–æ–ª—å) ->
                        #-> –¥–æ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∞—Å—Å–∏–≤–∞ (+2 –∏–∑-–∑–∞ —Ç–æ–≥–æ, —á—Ç–æ –¥–æ–±–∞–≤–∏–ª–∏ –ú)
    if i % 2 == 0: # –ï—Å–ª–∏ –Ω–æ–º–µ—Ä —ç–ª–µ–º–µ–Ω—Ç–∞ —á—ë—Ç–Ω—ã–π
        lamp_off = lamp_off + (program[i] - program[i-1]) # –ü—Ä–∏–±–∞–≤–ª—è–µ–º –≤—Ä–µ–º—è –Ω–µ—Ä–∞–±–æ—Ç–∞—é—â–µ–π –ª–∞–º–ø–æ—á–∫–∏
    else:
        lamp_on = lamp_on + (program[i] - program[i-1]) # –ü—Ä–∏–±–∞–≤–ª—è–µ–º –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç–∞—é—â–µ–π –ª–∞–º–ø–æ—á–∫–∏
maximum = lamp_on # –ü—Ä–∏–Ω–∏–º–∞–µ–º –∑–∞ –º–∞–∫—Å–∏–º—É–º –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –ª–∞–º–ø–æ—á–∫–∏

for i in range(1, n+1, 2): # –ü—Ä–æ–≤–µ—Ä—è–µ–º –ø–∞—Ä—ã —ç–ª–µ–º–µ–Ω—Ç–æ–≤
    final = final + (program[i] - program[i-1])
    if ((program[i] - program[i-1] > 1) or (program[i+1] - program[i] > 1)): # –ï—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –ø—Ä–æ—à–ª—ã–º —ç–ª–µ–º–µ–Ω—Ç–æ–º –∏–ª–∏ ->
                                                                 #-> –±—É–¥—É—â–µ–º —Ä–∞–≤–Ω–∞ 1, —Ç–æ –º–∞–∫—Å–∏–º—É–º –ø–µ—Ä—Å—á–∏—Ç—ã–≤–∞—Ç—å –Ω–µ –Ω—É–∂–Ω–æ ()
        maximum = max(maximum, final-1 + lamp_off) # –ó–∞ –Ω–æ–≤—ã–π –º–∞–∫—Å–∏–º—É–º –±–µ—Ä—ë–º –ª–∏–±–æ –ø—Ä–µ–∂–Ω–∏–π –º–∞–∫—Å–∏–º—É–º,->
        #-> –ª–∏–±–æ –Ω–∞–π–¥–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –ª–∞–º–ø–æ—á–∫–∏ + –æ—Å—Ç–∞–≤—à–µ–µ—Å—è –Ω–µ—Ä–∞–±–æ—á–µ–µ –≤—Ä–µ–º—è –ª–∞–º–ø–æ—á–∫–∏ –Ω–∞ —ç—Ç–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏
    lamp_off = lamp_off - (program[i+1] - program[i]) # –ü–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º –≤—Ä–µ–º—è –Ω–µ —Ä–∞–±–æ—Ç—ã –ª–∞–º–ø–æ—á–∫–∏

print("-----------------------------–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ----------------------------")
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ –≤—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è —Å–≤–µ—Ç–∞ = ", maximum) # –í—ã–≤–æ–¥ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ —Ä–∞–±–æ—Ç—ã –ª–∞–º–ø–æ—á–∫–∏
```

```python
print("-----------------------------–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ-----------------------------")
n, L, R, QL, QR = map(int, input("–í–≤–µ–¥–∏—Ç–µ n, L, R, QL, QR, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ: ").split()) # –í–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥–º–µ—Ç–æ–≤, 
# –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–µ–π—Å—Ç–≤–∏–π –ª–µ–≤–æ–π —Ä—É–∫–æ–π, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–µ–π—Å—Ç–≤–∏–π –ø—Ä–∞–≤–æ–π —Ä—É–∫–æ–π, —à—Ç—Ä–∞—Ñ –ª–µ–≤–æ–π —Ä—É–∫–∏, —à—Ç—Ä–∞—Ñ –ø—Ä–∞–≤–æ–π —Ä—É–∫–∏

W = list(map(int, input("–í–≤–µ–¥–∏—Ç–µ –≤–µ—Å–∞ –ø—Ä–µ–¥–º–µ—Ç–æ–≤: ").split())) # –°–ø–∏—Å–æ–∫ –≤–µ—Å–æ–≤ –ø—Ä–µ–¥–º–µ—Ç–æ–≤
 
sum_el = [0] # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–∞—Å—Å–∏–≤–∞ —Å—É–º–º–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –ø–µ—Ä–µ–Ω–æ—Å–∞ –ø—Ä–µ–¥–º–µ—Ç–æ–≤
for i in range(1, n + 1): # –û—Ç 1 –Ω–æ–º–µ—Ä–∞ —ç–ª–µ–º–µ–Ω—Ç–∞, –¥–æ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ (–∏–∑-–∑–∞ 0 –≤ –Ω–∞—á–∞–ª–µ —Å–º–µ—à–∞–µ–º –Ω–∞ +1)
    sum_el.append(W[i - 1] + sum_el[i - 1]) # –ó–∞–ø–æ–ª–Ω—è–µ–º —Å–ø–∏—Å–æ–∫ –ø—Ä–æ—Å—Å—É–º–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
 
answer = QR * (n - 1) + sum_el[n] * R # –†–∞—Å—á—ë—Ç –∑–∞—Ç—Ä–∞—á–µ–Ω–Ω–æ–π —ç–Ω–µ—Ä–≥–∏–∏ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –≤—Å—ë –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –ø—Ä–∞–≤–æ–π —Ä—É–∫–æ–π
 
for i in range(1, n + 1): # –ü—Ä–æ—Ö–æ–¥–∏–º —Ü–∏–∫–ª–æ–º –ø–æ —Å–ø–∏—Å–∫—É –ø—Ä–æ—Å—É–º–º–∏—Ä–æ–≤–∞–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
    energy = L * sum_el[i] + R * (sum_el[n] - sum_el[i]) # –†–∞—Å—á—ë—Ç —ç–Ω–µ—Ä–≥–∏–∏ –¥–æ —à—Ç—Ä–∞—Ñ–æ–≤ (–ø–µ—Ä–µ–±–æ—Ä —Ä—É–∫ –ø–æ —Å—É—Ç–∏)
    if i > (n - i): # –ï—Å–ª–∏ –ø–µ—Ä–µ–ª–æ–∂–∏–ª–∏ –ª–µ–≤–æ–π —Ä—É–∫–æ–π –±–æ–ª—å—à–µ —á–µ–º –ø—Ä–∞–≤–æ–π
        energy = energy + (i - (n - i) - 1) * QL # –°—á–∏—Ç–∞–µ–º —à—Ç—Ä–∞—Ñ –∑–∞ —ç—Ç–æ
    elif (n - i) > i: # –ò–Ω–∞—á–µ –µ—Å–ª–∏ –ø–µ—Ä–µ–ª–æ–∂–∏–ª–∏ –ø—Ä–∞–≤–æ–π —Ä—É–∫–æ–π –±–æ–ª—å—à–µ —á–µ–º –ª–µ–≤–æ–π
        energy = energy + ((n - i) - i - 1) * QR # –°—á–∏—Ç–∞–µ–º —à—Ç—Ä–∞—Ñ –∑–∞ —ç—Ç–æ
    if  energy < answer: # –ï—Å–ª–∏ —Ä–∞—Å—á–∏—Ç–∞–Ω–Ω–∞—è —ç–Ω–µ—Ä–≥–∏—è –º–µ–Ω—å—à–µ —Ä–∞–Ω–µ–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞, —Ç–æ —ç—Ç–∞ —ç–Ω–µ—Ä–≥–∏—è —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –æ—Ç–≤–µ—Ç–æ–º
        answer = energy

print("-----------------------------–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ----------------------------")
print("–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ –∑–∞—Ç—Ä–∞—á–µ–Ω–Ω–∞—è —ç–Ω–µ—Ä–≥–∏—è = ", answer) # –í—ã–≤–æ–¥–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—É—é —ç–Ω–µ—Ä–≥–∏—é
```

```python
print("-----------------------------–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ-----------------------------")
n = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–æ–Ω–µ—Ç: ")) # –í–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–æ–Ω–µ—Ç
arr = list(map(int, input("–í–≤–µ–¥–∏—Ç–µ –ø–æ–∑–∏—Ü–∏–∏ –º–æ–Ω–µ—Ç: ").split())) # –í–≤–æ–¥–∏–º –ø–æ–∑–∏—Ü–∏–∏ –º–æ–Ω–µ—Ç
pos = n # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –æ—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏—è –ø–æ–∑–∏—Ü–∏–∏ –º–æ–Ω–µ—Ç—ã
a = [0 for i in range(n+1)] # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –æ—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å –∑–∞–º–µ–Ω—É –º–æ–Ω–µ—Ç ->
                            #-> (0 - –≤—ã—à–ª–∞ –∏–∑ –æ–±—Ä–∞—â–µ–Ω–∏—è, 1 - –æ—Å—Ç–∞–ª–∞—Å—å –≤ –æ–±—Ä–∞—â–µ–Ω–∏–∏)
res = 1 # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –ø–æ–¥—Å—á—ë—Ç–∞ —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
ans = [1] # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥—É—Ç –∑–∞–ø–∏—Å–∞–Ω—ã —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è ->
          #-> (—Å—Ä–∞–∑—É –∑–∞–Ω–æ—Å–∏–º 1, —Ç–∞–∫ –∫–∞–∫ –∏–∑ —É—Å–ª–æ–≤–∏—è —Å–∫–∞–∑–∞–Ω–æ, —á—Ç–æ —Å–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞–≤–Ω–∞ 1)
for x in arr: # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–∑–∏—Ü–∏–π –º–æ–Ω–µ—Ç
    a[x] = 1 # –ü–æ–º–µ—á–∞–µ–º —Ç–µ–∫—É—â—É—é –º–æ–Ω–µ—Ç—É –∫–∞–∫ –∑–∞–º–µ–Ω—ë–Ω–Ω—É—é –Ω–∞ –∏—Å–ø–æ–ª—å–∑—É–µ–º—É—é –≤ –æ–±—Ä–∞—â–µ–Ω–∏–∏
    res = res + 1 # –ü–æ–≤—ã—à–∞–µ–º —Å–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–≤–∏–∞–Ω–∏—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
    while a[pos] == 1: # –ü–æ–∫–∞ —Ç–µ–∫—É—â–∞—è –º–æ–Ω–µ—Ç–∞ –Ω–∞ —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ –æ–±—Ä–∞—â–µ–Ω–∏–∏ ->
                       #-> (–ø–æ–∫–∞ –ø–æ—Å–ª–µ–¥–Ω—è—è –º–æ–Ω–µ—Ç–∞ –æ—Å—Ç–∞—ë—Ç—Å—è –≤—ã—à–µ–¥—à–µ–π –∏–∑ –æ–±—Ä–∞—â–µ–Ω–∏—è, —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ª–∏—à—å —Ä–∞—Å—Ç—ë—Ç)
        pos = pos - 1 # –ü—Ä–æ—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º, –±—ã–ª–∞ –ª–∏ –∑–∞–º–µ–Ω–µ–Ω–∞ –º–æ–Ω–µ—Ç–∞ —Å–ª–µ–≤–∞ ->
                #-> (—á–µ–º –±–ª–∏–∂–µ –∫ –Ω–∞—á–∞–ª—É –º—ã –≤—Å—Ç—Ä–µ—Ç–∏–º –≤—ã—à–µ–¥—à—É—é –º–æ–Ω–µ—Ç—É –∏–∑ –æ–±—Ä–∞—â–µ–Ω–∏—è, —Ç–µ–º –º–µ–Ω—å—à–µ –±—É–¥–µ—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è)
        res = res - 1 # –£–º–µ–Ω—å—à–∞–µ–º —Å–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è
    ans.append(res) # –î–æ–±–∞–≤–ª—è–µ–º —Å–ª–æ–∂–Ω–æ—Å—Ç—å –≤ —Å–ø–∏—Å–æ–∫ —Å–ª–æ–∂–Ω–æ—Å—Ç–µ–π —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è
print("-----------------------------–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ----------------------------")
print("–°–ª–æ–∂–Ω–æ—Å—Ç—å —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ—Å–ª–µ –∑–∞–º–µ–Ω—ã –º–æ–Ω–µ—Ç—ã:", ' '.join(map(str, ans))) # –í—ã–≤–æ–¥–∏–º –æ—Ç–≤–µ—Ç
```
